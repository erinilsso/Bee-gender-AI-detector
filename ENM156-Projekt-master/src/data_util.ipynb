```python
import os
from xml.dom import minidom

label_dirs = [
  '../data/Beehive2_summer/labels/yolo', 
  '../data/Raw data/Pictures Ericsson/Pictures/labels/yolo', 
  #'../data/Raw data/Pictures LE/pictures_hive1/labels/yolo',
  '../data/Raw data/Pictures LE/pictures_hive2/labels/yolo',
]
classes = ['Worker/female', 'Drone/male', 'Unknown/bee']
```

```python
# Find number of labeled images
tot = 0
for dir in label_dirs:
  count = len(os.listdir(dir))-1  # -1 for 'yolo' folder
  tot += count
  print(str(count) + " (" + dir + ")")
print("Total: " + str(tot))
```

```python
# Find labels marked as difficult (not including 'hive2')
tot_diff_count = 0
for label_dir in label_dirs:
  dir_diff_count = 0
  for filename in os.listdir(label_dir):
    if os.path.splitext(filename)[1] != '.xml':
      continue
    filepath = os.path.join(label_dir, filename)
    dom = minidom.parse(filepath)
    difficult = dom.getElementsByTagName('difficult')
    diff_count = 0
    for d in difficult:
      if d.childNodes[0].nodeValue == '1':
        diff_count += 1
    if diff_count > 0:
      print(filename + ": " + str(diff_count))
    dir_diff_count += diff_count
  tot_diff_count += dir_diff_count
  print(str(dir_diff_count) + " (" + label_dir + ")")
print("Total: " + str(tot_diff_count))
```

```python
# Find number of samples per class/label
total_class_counts = dict.fromkeys(classes, 0)
for label_dir in label_dirs:
  print(label_dir)
  dir_class_counts = dict.fromkeys(classes, 0)
  for filename in os.listdir(label_dir):
    if os.path.splitext(filename)[1] != '.xml':
      continue
    filepath = os.path.join(label_dir, filename)
    dom = minidom.parse(filepath)
    labels = dom.getElementsByTagName('name')
    for label in labels:
      total_class_counts[label.childNodes[0].nodeValue] += 1
      dir_class_counts[label.childNodes[0].nodeValue] += 1
  print(dir_class_counts)

print("Total:")
print(total_class_counts)
```

```python
# Find files with labels
files = { c: [] for c in classes }
for label_dir in label_dirs:
  for filename in os.listdir(label_dir):
    if os.path.splitext(filename)[1] != '.xml':
      continue
    found_class = dict.fromkeys(classes, False)
    filepath = os.path.join(label_dir, filename)
    dom = minidom.parse(filepath)
    labels = dom.getElementsByTagName('name')
    for label in labels:
      if found_class[label.childNodes[0].nodeValue] == False:
        files[label.childNodes[0].nodeValue].append(filepath)
        found_class[label.childNodes[0].nodeValue] = True

print(files)
```

```python
# Find ratio of labels to images
label_count = 0
img_count = 0
for label_dir in label_dirs:
  for label_file in os.listdir(label_dir):
    label_count += 1
  for img_file in os.listdir(os.path.dirname(label_dir)):
    if os.path.splitext(img_file)[1] == '.jpg' or os.path.splitext(img_file)[1] == '.jpeg':
      img_count += 1

print(str(label_count) + " of " + str(img_count) + " images are labeled (" + str(100*label_count/img_count) + "%)")
```
